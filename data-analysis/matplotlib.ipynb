```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot with default settings.
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosine = np.cos(x)
sine = np.sin(x)

plot(x, cosine)
plot(x, sine)
```

```python
# Create a figure of size 8 x 6 inches, 80 dots per inch.
figure(figsize=(8, 6), dpi=80)

# Create a new subplot from a grid of 1x1
subplot(1, 1, 1)

# Plot cosine with a blue continuous line of width 2
plot(x, cosine, color='blue', linewidth=2, linestyle='-', label='cosine')

# Plot sine with a red continuous line of width 2
plot(x, sine, color='red', linewidth=2, linestyle='-', label='sine')

margin = 1.1

# Set x limits
xlim(x.min()*margin, x.max()*margin)

# Set x ticks, and using Latex to label x ticks.
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])

# Set y limits
ylim(cosine.min()*margin, cosine.max()*margin)

# Set y ticks
yticks(np.linspace(-1, 1, 5, endpoint=True))

legend(loc='upper left')
```

```python
# Scatter plots
n = 1024
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)
angle = np.arctan2(y, x)
scatter(x, y, s=75, c=angle, cmap=matplotlib.cm.cool, alpha=.5)
```

```python
# Bar Plots
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')

ylim(-1.25, +1.25)
```
